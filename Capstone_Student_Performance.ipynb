```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Utkarsh Internship/mat2.csv')
```

```python
df.head()
```

```python
df.drop('unnamed',axis=1,inplace=True)
```

```python
df.head()
```

```python
df_filtered = df[df['G3'] > 15]
df_filtered.head()
```

```python
average_grades = df.groupby('school')['G3'].mean()
print(average_grades)
```

```python
pivot_table = df.pivot_table(values='G3', index='sex', columns='school', aggfunc='mean')
print(pivot_table)
```

```python
sns.pointplot(x='school', y='G3', hue='sex', data=df, dodge=True)
plt.xlabel('School')
plt.ylabel('Average Final Grade')
plt.title('Average Final Grade by Sex and School')
plt.show()
```

```python
sns.jointplot(x='studytime', y='G3', data=df)
plt.show()
```

```python
sns.pairplot(df[['G1', 'G2', 'G3']])
plt.show()
```

```python
sns.catplot(x='romantic', y='G3', kind='violin', data=df)
plt.xlabel('Romantic Relationship Status')
plt.ylabel('Final Grade')
plt.title('Final Grade by Romantic Relationship Status')
plt.show()
```

```python
sns.displot(df['G3'], kde=True)
plt.xlabel('Final Grade')
plt.ylabel('Density')
plt.title('Distribution of Final Grades')
plt.show()
```

```python
attributes = df.columns

class Student:
    def __init__(self, attributes):
      for att in attributes:
        self.att = att

    def calculate_average_grade(self):
        average = (self.g1 + self.g2 + self.g3) / 3
        return average
```
